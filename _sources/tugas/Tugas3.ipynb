**Langkah 1** 

Langkah pertama yaitu mengimpor library yang diperlukan.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Langkah 2**

Langkah kedua adalah mengimpor data Iris. Setelah itu menetapkan 4 variabel independen ke nilai X dan variabel dependen "Spesies" ke nilai Y. 

```python
dataset = pd.read_csv('https://gist.githubusercontent.com/hilmifa23/188a63a098ee77dd94951bfd1e0f8deb/raw/4c7d12a8d535a72298b2868ae437614ea717e4d8/IrisDataset')
X = dataset.iloc[:,:4].values
y = dataset['species'].values
dataset.head(5)
```

**Langkah 3**

Langkah ketiga adalah memisahkan data tersebut menjadi set pelatihan dan set pengujian. Digunakan nilai test_size=0,2 yang artinya 20% dari dataset akan diambil sebagai set pengujian dan 80% sisanya akan digunakan sebagai set pelatihan.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

**Langkah 4**

Langkah keempat dataset diubah menjadi rentang yang lebih kecil dengan menggunakan opsi Feature Scaling.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Langkah 5**

Pada langkah ini, digunakan kelas GaussianNB dari library sklearn.naive_bayes. Dalam kasus ini, digunakan model Gaussian. Untuk melatih model, kita menugaskan kelas GaussianNB ke variabel classifier dan melakukan pelatihan dengan memasukkan nilai X_train dan y_train ke dalamnya.

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

**Langkah 6**

Langkah keenam adalah melakukan prediksi nilai pada set pengujian menggunakan classifier.predict(), dan hasil prediksi akan disimpan ke dalam variabel y_pred.

```python
y_pred = classifier.predict(X_test) 
y_pred
```

**Langkah 7**

Pada langkah ini, biasanya digunakan dalam teknik klasifikasi untuk mengevaluasi akurasi model yang telah dilatih dan memvisualisasikan confusion matrix. Confusion matrix adalah sebuah tabel yang digunakan untuk menunjukkan jumlah prediksi yang benar dan salah pada masalah klasifikasi ketika nilai sebenarnya dari set pengujian diketahui. Tabel ini terdiri dari empat sel, yaitu True Positive (TP), False Positive (FP), True Negative (TN), dan False Negative (FN). True Positive (TP) merupakan jumlah prediksi yang benar dari kelas positif, sedangkan False Positive (FP) merupakan jumlah prediksi yang salah dari kelas negatif. True Negative (TN) merupakan jumlah prediksi yang benar dari kelas negatif, dan False Negative (FN) merupakan jumlah prediksi yang salah dari kelas positif.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

Berdasarkan confusion matrix di atas, dapat disimpulkan bahwa dari total 30 data pada set pengujian, sebanyak 29 data diklasifikasikan dengan benar dan hanya 1 data yang salah diklasifikasikan. Hal ini menghasilkan akurasi yang tinggi yaitu sebesar 96,67%.

**Langkah 8**

Pada tahap ini, dibuat sebuah Pandas DataFrame untuk membandingkan nilai-nilai klasifikasi antara data asli pada set pengujian (y_test) dan hasil prediksi (y_pred). Tahap ini merupakan langkah tambahan yang tidak memberikan informasi banyak seperti Confusion Matrix dan biasanya digunakan dalam regresi untuk memeriksa akurasi nilai prediksi. Seperti yang terlihat pada hasil tersebut, terdapat satu prediksi yang salah yang memprediksi spesies versicolor sebagai virginica. Namun demikian, secara keseluruhan, model Naive Bayes untuk klasifikasi yang telah dilatih memiliki performa yang sangat baik, dengan tingkat akurasi mencapai 96,67%.
